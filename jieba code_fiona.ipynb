#Combine multiple files into one file

```python
# -*- coding:utf-8 -*-

import os 
path = u'C:\\Users\\BigData\\Desktop\\project\\JY'

fname = u'combine_jy.txt'


import codecs

with codecs.open(os.path.join(os.environ['USERPROFILE'],fname ),'a','utf-8') as total :
    for filename in os.listdir(path) :
        fullpath = os.path.join(path,filename)
        with codecs.open(fullpath,'r','utf-8') as f:
            total.write(f.read())
            total.write(u'\r\n')
            f.close()
    total.close()
```

#Jieba 斷詞 + 濾詞

```python
import re
import codecs
import jieba

filename = 'combine_jy.txt'
f = codecs.open(filename,'r','utf-8')
tmp = f.read()

text = re.sub(u'[\s, ，\.。:\?？\!！「」 ─┐ ─┘ \[ \] \│ \── \─┼ \┼─ \├─ \├ \├─ \┤ \= \┬┬ \┴┴ \d \(\) （）　： 【】 、 ； ○ ----- ]',' ',tmp)
text = re.sub('[ ]{2,}' ,' ',text)
f.close()

filename = 'clean for Jieba.txt'
f = codecs.open(filename,'w','utf-8')
f.write(text)

f.close()


jieba.load_userdict("0627mix_ver.txt")
content = open('clean for Jieba.txt', 'rb').read()
words = jieba.cut(content, cut_all=False)

dic_list = []
stopword = open('0629UselessDict.txt', 'rb')
for ele in stopword:
    if ele not in dic_list:
        dic_list.append(ele.strip().decode('utf-8'))

stopword.close()

with codecs.open('JiebaResult.txt','w','utf-8') as jr :
    for word in words:
        if word not in dic_list:
            jr.write(word + u'\r\n')
    
    jr.close()
    
    
```

#計算詞頻

```python
import codecs
import operator

srcpath = 'JiebaResult.txt'

w_dic = {}
with codecs.open(srcpath,'r','utf-8') as src :
    for word in src.readlines() :
        word = word.strip()
        if word != '' :
            if not word in w_dic :
                w_dic[word] = 1
            else :
                w_dic[word] += 1
    src.close()

    
words_freq = sorted(w_dic.iteritems(),key=operator.itemgetter(1),reverse=True)
    
for word in words_freq:
    if word[1] >= 2:
        print word[0], word[1]
```
